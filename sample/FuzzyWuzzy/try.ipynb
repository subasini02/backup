```python
from fuzzywuzzy import fuzz
```

## fuzz.ratio:
calc the distance between some ordering of the token both input string.

```python
fuzz.ratio("maheshwaran","maheshwari")
```

## fuzz.partial_ratio:
finds the shorter string and its lenght. Matches it with all sub-strings of same length then give the ratio.

```python
fuzz.partial_ratio("Boggarapu M. Yaswanth","Boggarapu Yaswanth")
```

## fuzz.token_sort_ratio:

account for similar strings that are out of order

```python
fuzz.token_sort_ratio("maheshwaran","maheshwari")
```

# RapidFuzz

```python
! pip install rapidfuzz
```

```python
from rapidfuzz import fuzz
```

## rapidfuzz.fuzz.ratio

Calculates the normalized Indel distance.

```python
fuzz.ratio("Boggarapu M. Yaswanth","Boggarapu Yaswanth")
```

## rapidfuzz.fuzz.partial_ratio:

Searches for the optimal alignment of the shorter string in the longer string and returns the fuzz.ratio for this alignment.

```python
fuzz.partial_ratio("s bharath","siddtharth bharath")
```

## rapidfuzz.fuzz.partial_ratio_alignment

Searches for the optimal alignment of the shorter string in the longer string and returns the fuzz.ratio and the corresponding alignment.

```python
fuzz.partial_ratio_alignment("Boggarapu M. Yaswanth","Boggarapu Yaswanth")
```

## rapidfuzz.fuzz.token_set_ratio

Compares the words in the strings based on unique and common words between them using fuzz.ratio

```python
fuzz.token_set_ratio("Boggarapu M. Yaswanth","Boggarapu Yaswanth")
```

## rapidfuzz.fuzz.partial_token_set_ratio

Compares the words in the strings based on unique and common words between them using fuzz.partial_ratio

```python
fuzz.partial_token_set_ratio("Boggarapu M. Yaswanth","Boggarapu Yaswanth")
```

```python
import pandas as pd
from fuzzywuzzy import fuzz

data = pd.read_csv("./sorted_file_1.csv")

def calculate_similarity(row):
    return fuzz.ratio(str(row['aggr_full_name_initial']), str(row['full_name']))

data['sim_sco_full_name'] = data.apply(calculate_similarity, axis=1)

def calculate_similarity_1(row):
    return fuzz.ratio(str(row['aggr_full_name_initial']), str(row['first_name']))

data['sim_sco_first_name'] = data.apply(calculate_similarity_1, axis=1)

def calculate_similarity_2(row):
    return fuzz.ratio(str(row['aggr_full_name_initial']), str(row['last_name']))

data['sim_sco_last_name'] = data.apply(calculate_similarity_2, axis=1)

data.to_csv("./new.csv",index=False)

```
