```python
import pandas as pd

data = pd.read_csv("./sorted_file_1.csv")
df = pd.DataFrame(data)

# Function to extract first name only and remove single characters and '.'
def extract_firstname(name):
    # Split the name by whitespace
    parts = name.split()
    # Filter out single characters and '.' from the parts
    filtered_parts = [part for part in parts if len(part) > 1 and part != '.']
    # Return the joined filtered parts
    return ' '.join(filtered_parts)

# Apply the function to 'aggr_full_name' column
df['aggr_full_name_initial'] = df['aggr_full_name'].apply(extract_firstname)

# Save DataFrame to a new CSV file
df.to_csv("./sorted_file_1.csv", index=False)
```

```python
import pandas as pd

# Read data from CSV file
df = pd.read_csv("./sorted_file_1.csv")

# Function to split full name into first name and last name
def split_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = ' '.join(parts[1:]) if len(parts) > 1 else ''  # Join remaining parts as last name
    return first_name, last_name

# Apply the function to 'aggr_full_name' column and store in separate columns
df['first_name'], df['last_name'] = zip(*df['full_name'].map(split_name))


# Save DataFrame to a new CSV file
df.to_csv("./sorted_file_1.csv", index=False)
```
