```python
import pandas as pd

# Read data from CSV file
# df = pd.read_csv("./cleaned_names.csv")

# Function to split full name into first name and last name
def split_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = ' '.join(parts[1:]) if len(parts) > 1 else ''  # Join remaining parts as last name
    return first_name, last_name

# Apply the function to 'aggr_full_name' column and store in separate columns
df['first_name'], df['last_name'] = zip(*df['full_name'].map(split_name))

df['agg_first_name'], df['agg_last_name'] = zip(*df['aggr_full_name'].map(split_name))


# Save DataFrame to a new CSV file
df.to_csv("./new.csv", index=False)
```

```python
import pandas as pd
from fuzzywuzzy import fuzz

data = pd.read_csv("./new.csv")

data['aggr_full_name'] = data['aggr_full_name'].str.lower()
data['full_name'] = data['full_name'].str.lower()
data['first_name'] = data['first_name'].str.lower()
data['last_name'] = data['last_name'].str.lower()
data['agg_first_name'] = data['agg_first_name'].str.lower()
data['agg_last_name'] = data['agg_last_name'].str.lower()

def calculate_similarity(row):
    return fuzz.ratio(str(row['aggr_full_name']), str(row['full_name']))

data['sim_sco_full_name'] = data.apply(calculate_similarity, axis=1)

def calculate_similarity_1(row):
    return fuzz.ratio(str(row['agg_first_name']), str(row['first_name']))

data['sim_sco_first_name'] = data.apply(calculate_similarity_1, axis=1)

def calculate_similarity_2(row):
    return fuzz.ratio(str(row['agg_last_name']), str(row['last_name']))

data['sim_sco_last_name'] = data.apply(calculate_similarity_2, axis=1)

data.to_csv("./new.csv",index=False)

```
